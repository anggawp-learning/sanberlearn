```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(ncols=3, sharex=True, sharey=True, figsize=(20,8))
ax[0].plot(x, tan, color='xkcd:blue', label='tanX')
ax[1].plot(x, sin, color='xkcd:red', label='sinX')
ax[2].plot(x, cos, color='xkcd:green', label='cosX')
ax[0].set_title('Grafik tanX')
ax[1].set_title('Grafik sinX')
ax[2].set_title('Grafik cosX')
fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(16,10))
ax.plot(x,y)
ax.set_title('Grafik CosX')
in2 = ax.inset_axes([-0.1,-0.1,0.3,0.3])
in2.plot(x,y)
in2.plot(x,y2,color='xkcd:red')
in2.set_title('Grafik CosX^2')
in3 = ax.inset_axes([0.9,-0.1,0.3,0.3])
in3.plot(x,y)
in3.plot(x,y3, color='xkcd:green')
in3.set_title('Grafik CosX^3')
in4 = ax.inset_axes([0.9, 0.9,0.3,0.3])
in4.plot(x,y)
in4.plot(x,y4, color='xkcd:almost black')
in4.set_title('Grafik CosX^4')
in5 = ax.inset_axes([-0.1, 0.9,0.3,0.3])
in5.plot(x,y)
in5.plot(x,y5, color='xkcd:green apple')
in5.set_title('Grafik CosX^5')

plt.show()
```
