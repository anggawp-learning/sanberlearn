```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(figsize = (30,12), ncols=3,nrows=1,sharex=True,sharey=True)
ax[0].plot(x,tan,color='blue',label = 'tanX')
ax[0].set_title('Grafik tanx')
ax[1].plot(x,sin,color='red',label = 'sinX')
ax[1].set_title('Grafik sinx')
ax[2].plot(x,cos,color='green',label = 'cosX')
ax[2].set_title('Grafik cosx')
lines_labels = [ax.get_legend_handles_labels() for ax in fig.axes]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
fig.legend(lines, labels,loc="upper center",fontsize=15)
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(x,y)
ins1 = ax.inset_axes([-0.1,0.9,0.3,0.3])
ins1.plot(x,y2)
plt.show()
```
