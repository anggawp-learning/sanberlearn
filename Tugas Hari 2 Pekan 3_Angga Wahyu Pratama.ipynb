Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
import matplotlib.pylab as plt
fig, ax = plt.subplots(figsize=(10,6))

ax.plot(x,cos,color='red',label='Cosx')
ax.plot(x,cos3,color='green',linestyle='--',label='3 Cosx')
ax.plot(x,sin,color='blue',linestyle='dashdot', label='Sin(x-90)')
ax.plot(x,sin2,color='darkkhaki',linestyle='',marker='x',label='2 Sinx')
ax.set_title('Grafik Cos Sin')
ax.legend(loc='center',fontsize=9)
plt.show()
```
